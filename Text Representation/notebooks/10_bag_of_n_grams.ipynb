# Bag of N-Grams - Capturing Word Sequences

## Overview

**N-grams** extend the Bag of Words concept by considering sequences of N consecutive words instead of individual words. This helps capture some context and word order that BoW misses.

### Types of N-Grams

| N | Name | Example from "I love machine learning" |
|:-:|:-----|:---------------------------------------|
| 1 | Unigram | "I", "love", "machine", "learning" |
| 2 | Bigram | "I love", "love machine", "machine learning" |
| 3 | Trigram | "I love machine", "love machine learning" |

### Why N-Grams Matter

Consider: *"The movie was **not good**"* vs *"The movie was **good**"*

- **Unigrams**: Both have "good" ‚Üí Can't distinguish sentiment!
- **Bigrams**: "not good" vs "was good" ‚Üí Can distinguish!

## üîß Generating N-Grams with CountVectorizer

The `ngram_range` parameter controls which n-grams to include:
- `ngram_range=(1, 1)` ‚Üí Unigrams only (standard BoW)
- `ngram_range=(1, 2)` ‚Üí Unigrams + Bigrams
- `ngram_range=(2, 2)` ‚Üí Bigrams only
- `ngram_range=(1, 3)` ‚Üí Unigrams + Bigrams + Trigrams

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
v.fit(["Thor Hathodawala is looking for a job"])
v.vocabulary_
```

### Unigrams Only (Default)

```python
v = CountVectorizer(ngram_range=(1,2))
v.fit(["Thor Hathodawala is looking for a job"])
v.vocabulary_
```

### Adding Bigrams (1, 2)

Notice how the vocabulary now includes word pairs like "thor hathodawala", "is looking", etc.

```python
v = CountVectorizer(ngram_range=(1,3))
v.fit(["Thor Hathodawala is looking for a job"])
v.vocabulary_
```

### Adding Trigrams (1, 3)

Now we also have 3-word sequences. Notice how vocabulary size grows significantly!

## üìù Text Preprocessing with spaCy

Before creating n-grams, we should preprocess text to:
1. **Remove stop words** ("the", "is", "a") - they don't add meaning
2. **Lemmatize** - reduce words to base form ("eating" ‚Üí "eat")
3. **Remove punctuation** - not useful for classification

This reduces vocabulary size and focuses on meaningful terms.

```python
corpus = [
    "Thor ate pizza",
    "Loki is tall",
    "Loki is eating pizza"
]
```

```python
import spacy

# load english language model and create nlp object from it
nlp = spacy.load("en_core_web_sm") 

def preprocess(text):
    # remove stop words and lemmatize the text
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    
    return " ".join(filtered_tokens) 
```

```python
preprocess("Thor ate pizza")
```

```python
preprocess("Loki is eating pizza")
```

Notice how:
- "is" (stop word) is removed
- "eating" ‚Üí "eat" (lemmatized)

```python
corpus_processed = [
    preprocess(text) for text in corpus
]
corpus_processed
```

```python
v = CountVectorizer(ngram_range=(1,2))
v.fit(corpus_processed)
v.vocabulary_
```

### Generating Vectors for Documents

Now let's create n-gram vectors for sample documents:

```python
v.transform(["Thor eat pizza"]).toarray()
```

### Handling Out-of-Vocabulary (OOV) Words

What happens when we encounter a word not in our vocabulary?

```python
v.transform(["Hulk eat pizza"]).toarray()
```

"Hulk" wasn't in training data, so it's ignored. Only "eat pizza" is captured.

‚ö†Ô∏è **Limitation**: N-grams cannot represent words/phrases not seen during training.

## ‚öñÔ∏è Handling Class Imbalance

The SCIENCE category has ~1/3rd the samples of other categories. This can cause the model to:
- Predict majority classes more often
- Have lower performance on minority classes

### Techniques to Handle Imbalance:

| Technique | Description |
|:----------|:------------|
| **Undersampling** | Reduce majority class samples |
| **Oversampling** | Duplicate minority class samples |
| **SMOTE** | Generate synthetic minority samples |
| **Class weights** | Penalize misclassifying minority class more |

We'll use **undersampling** for simplicity (though SMOTE is preferred in production).

```python
min_samples = 1381 # we have these many SCIENCE articles and SCIENCE is our minority class


df_business = df[df.category=="BUSINESS"].sample(min_samples, random_state=2022)
df_sports = df[df.category=="SPORTS"].sample(min_samples, random_state=2022)
df_crime = df[df.category=="CRIME"].sample(min_samples, random_state=2022)
df_science = df[df.category=="SCIENCE"].sample(min_samples, random_state=2022)
```

```python
df_balanced = pd.concat([df_business,df_sports,df_crime,df_science],axis=0)
df_balanced.category.value_counts()
```

**Convert text category to a number**

```python
target = {'BUSINESS': 0, 'SPORTS': 1, 'CRIME': 2, 'SCIENCE': 3}

df_balanced['category_num'] = df_balanced['category'].map({
    'BUSINESS': 0,
    'SPORTS': 1, 
    'CRIME': 2, 
    'SCIENCE': 3
})
```

```python
df_balanced.head()
```

Now all categories have equal samples (1381 each), creating a balanced dataset.

## üß™ Experiment 1: Raw Text (No Preprocessing)

First, let's train models on raw text to establish a baseline. We'll compare different n-gram ranges.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_balanced.text, 
    df_balanced.category_num, 
    test_size=0.2, # 20% samples will go to test dataset
    random_state=2022,
    stratify=df_balanced.category_num
)
```

```python
print(X_train.shape)
X_train.head()
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### Attempt 1: Unigrams Only (1, 1) - Standard BoW

Using only single words (no word sequences):

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_bow', CountVectorizer(ngram_range = (1, 1))),        #using the ngram_range parameter 
     ('Multi NB', MultinomialNB())         
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

### Attempt 2: Unigrams + Bigrams (1, 2)

Adding word pairs to capture some context:

```python
#1. create a pipeline object
clf = Pipeline([
    ('vectorizer_1_2_gram', CountVectorizer(ngram_range = (1, 2))),        #using the ngram_range parameter 
     ('Multi NB', MultinomialNB())         
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

### Attempt 3: Unigrams + Bigrams + Trigrams (1, 3)

Adding 3-word sequences for even more context:

```python
#1. create a pipeline object
clf = Pipeline([
    ('vectorizer_1_3_grams', CountVectorizer(ngram_range = (1, 3))),        #using the ngram_range parameter 
     ('Multi NB', MultinomialNB())         
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

## üéØ Key Takeaways

### N-Gram Trade-offs

| N-gram Range | Pros | Cons |
|:-------------|:-----|:-----|
| (1, 1) Unigrams | Small vocabulary, fast | No context |
| (1, 2) + Bigrams | Captures phrases, good balance | Larger vocabulary |
| (1, 3) + Trigrams | More context | Much larger vocabulary, may overfit |

### Best Practices

1. **Start with (1, 2)** - usually the best trade-off
2. **Preprocess text** - removes noise, improves performance
3. **Handle class imbalance** - use SMOTE or class weights in production
4. **Use `max_features`** - limit vocabulary size to prevent overfitting
